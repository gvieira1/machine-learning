```python
mpg_1 = mpg.drop('name', axis=1)
```

```python
mpg_2 = mpg_1.dropna()
```

```python
mpg_2
```

```python
one_hot = pd.get_dummies(mpg_2, columns=['origin'])
```

```python
one_hot
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
normalizar = one_hot.drop(['origin_europe', 'origin_japan', 'origin_usa', 'model_year', 'mpg'], axis=1)
one_hot_encoded = one_hot[['origin_europe', 'origin_japan', 'origin_usa', 'model_year', 'mpg']]
```

```python
df_normalizado = scaler.fit_transform(normalizar)
df_normalizado = pd.DataFrame(df_normalizado, columns=normalizar.columns)
df_final = pd.concat([df_normalizado, one_hot_encoded], axis=1)
```

```python
df_final = df_final.dropna()
```

```python
df_final
```

```python
from sklearn.model_selection import train_test_split

x = df_final.drop(['mpg'], axis=1)
y = df_final['mpg']

X_treino, X_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
```

```python
#BaggingRegressor
from sklearn.ensemble import BaggingRegressor
```

```python
dt = DecisionTreeRegressor(max_depth=4)
```

```python
bagging_reg = BaggingRegressor(estimator=dt)
```

```python
bagging_reg.fit(X_treino, y_treino)
```

```python
predictions = bagging_reg.predict(X_teste)
```

```python
print("BaggingRegressor")
print("")
print("RMSE: ", np.sqrt(mean_squared_error(y_teste, predictions)))
print("")
r2 = r2_score(y_teste, predictions)
print("R²: ", r2)
```

```python
#Boosting
from sklearn.ensemble import AdaBoostRegressor
```

```python
boosting = AdaBoostRegressor(estimator=dt)
```

```python
boosting.fit(X_treino, y_treino)
```

```python
predictions_ada = boosting.predict(X_teste)
```

```python
print("AdaBoostRegressor")
print("")
print("RMSE: ", np.sqrt(mean_squared_error(y_teste, predictions_ada)))
print("")
r2 = r2_score(y_teste, predictions_ada)
print("R²: ", r2)
```

```python
#Stacking
from sklearn.ensemble import StackingRegressor
```

```python
sv_reg = SVR()
linear = LinearRegression()
```

```python
models = [("lr", linear), ("dt", dt), ("SVR", sv_reg)]
```

```python
meta_learner = LinearRegression()
```

```python
s_reg = StackingRegressor(estimators=models, final_estimator=meta_learner)
```

```python
s_reg.fit(X_treino, y_treino)
```

```python
predictions_st = s_reg.predict(X_teste)
```

```python
print("StackingRegressor")
print("")
print("RMSE: ", np.sqrt(mean_squared_error(y_teste, predictions_st)))
print("")
r2 = r2_score(y_teste, predictions_st)
print("R²: ", r2)
```

------------------------------------------

-RMSE:  4.580 --> menor RMSE

StackingRegressor:

-RMSE:  4.648

-R²:  0.664

Depois dos testes, temos que o modelo com o melhor desempenho foi o AdaBoost seguido pelo StackingRegressor. Isso é esperado, já que o AdaBoost ajusta os erros do modelo anterior, permitindo um aprendizado mais eficiente das interações entre as variáveis, apesar de ser mais sensível a outliers. 

O resultado do StackingRegressor indica que a combinação de modelos não foi tão eficaz nesse contexto específico, que é uma de suas desvantagens.

O pior desempenho ficou com o Bagging, uma vez que ele depende de modelos de base simples (geralmente árvores de decisão rasas), podendo não ser capaz de capturar relações complexas entre as variáveis.
