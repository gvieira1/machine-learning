```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('new_cars.csv')
```

```python
df
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
y = df['mpg']
X = df.drop(['mpg'], axis=1)
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
LR = LinearRegression()
```

```python
LR.fit(X_treino, y_treino)
```

```python
predictions = LR.predict(X_teste)
```

```python
mae = mean_absolute_error(y_true=y_teste, y_pred=predictions)
mse = mean_squared_error(y_true=y_teste, y_pred=predictions)
rmse = root_mean_squared_error(y_true=y_teste, y_pred=predictions)
r2 = r2_score(y_true=y_teste, y_pred=predictions)
```

```python
print('LinearRegression()')
print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
print('R2: ', r2)
```
