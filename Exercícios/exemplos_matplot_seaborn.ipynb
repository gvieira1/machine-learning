```python
X = np.linspace(-np.pi, np.pi, 15)
S = np.sin(X)
C = np.cos(X)
ax = plt.axes([0.1,0.1,0.8,0.8])

ax1 = ax.plot(X, C, 'bs:')
ax2 = ax.plot(X, S, 'ro:')
ax.legend(labels=('Cosseno', 'Seno'), loc='upper left')
ax.set_title('Função Seno e Cosseno')
plt.show()
```

```python
import seaborn as sns
gorjetas = sns.load_dataset('tips')
```

```python
gorjetas
```

```python
sns.histplot(gorjetas['total_bill'], kde=True)
```

```python
sns.jointplot(gorjetas, x='total_bill', y='tip', kind='resid')
```

```python
sns.jointplot(gorjetas, x='total_bill', y='tip', hue='smoker')
```

```python
sns.pairplot(gorjetas, hue='smoker', kind='hist')
```

```python
sns.kdeplot(gorjetas['total_bill'])
sns.rugplot(gorjetas['total_bill'])
```

```python
sns.rugplot(gorjetas['total_bill'])
```

```python
sns.barplot( x='smoker', y='total_bill', data=gorjetas, fill=False)
```

```python
sns.countplot(x='smoker', data=gorjetas)
```

```python
sns.boxplot(x='day', y='total_bill', data=gorjetas)
```

```python
sns.boxplot(data=gorjetas, orient='h')
```

```python
sns.violinplot(x='day', y='total_bill', data=gorjetas)
```

```python
sns.violinplot(x='day', y='total_bill', data=gorjetas, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=gorjetas, hue='smoker', split=True)
```

```python
sns.stripplot(data=gorjetas, x='day', y='total_bill', hue='smoker')
```

```python
sns.swarmplot(data=gorjetas, x='day', y='total_bill', hue='smoker')
```

```python
sns.violinplot(data=gorjetas, x='day', y='total_bill')
sns.swarmplot(data=gorjetas, x='day', y='total_bill')
```
